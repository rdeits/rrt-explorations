```python
import random
import numpy as np
from matplotlib.path import Path
from IPython.display import clear_output
```

```python
class PolygonalObstacle2D:
    def __init__(self, xs, ys):
        if xs[-1] != xs[0] or ys[-1] != ys[0]:
            # ensure closed path
            xs = np.hstack((xs, xs[0]))
            ys = np.hstack((ys, ys[0]))
            
        self.path = Path(np.vstack((xs, ys)).T)
        self.xs = xs
        self.ys = ys
        
    def constraint(self, point):
        return self.path.contains_point(point)
    
    def draw(self, ax, *args, **kwargs):
        ax.plot(self.xs, self.ys, *args, **kwargs)
        
obstacles = [PolygonalObstacle2D([1,3,3,1], [-3, -3, 3, 3]),
             PolygonalObstacle2D([-2, 1, 1, -2], [-3, -3, -2, -2]),
             PolygonalObstacle2D([-2, 1, 1, -2], [2, 2, 3, 3]),
             PolygonalObstacle2D([5, 6, 6, 5], [1, 1, 10, 10]),
             PolygonalObstacle2D([5, 6, 6, 5], [-1, -1, -10, -10])]
```

```python
x_lim = [-10, 10]
y_lim = [-10, 10]

feas_map = np.ones((100,100))
# feas_map[45:65, 70:78] = 0
xy2px = np.array([[5, 0, 50],
                  [0, 5, 50]])
px2xy = np.array([[0.2, 0, -10],
                  [0, 0.2, -10]])
for px_x, row in enumerate(feas_map):
    for px_y, e in enumerate(row):
        xy = px2xy.dot(np.hstack((px_x, px_y, 1)))
        for o in obstacles:
            if o.constraint(xy):
                feas_map[px_x, px_y] = 0
                break


def distance(p0, p1):
    return np.sqrt(np.sum(np.power(np.subtract(p0, p1), 2)))

def random_state_xy():
    xy = np.array([random.uniform(*x_lim), random.uniform(*y_lim)])
    return xy

def closest_node(point, nodes):
    min_dist = np.inf
    best_node = None
    for node in nodes:
        d = node.dist_to(point)
        if d < min_dist:
            best_node = node
            min_dist = d
    return best_node


def state_check(point):
    px = xy2px.dot(np.hstack((point, 1)))
    return feas_map[px[0], px[1]]
```

```python
class Tree(object):
    def __init__(self, point, children=None, parent=None):
        self.point = np.array(point)
        
        if children is None:
            children = [] # unsafe to set this default in args, since appending to it later will affect the default for all future calls
            
        self.children = children
        self.parent = parent
        
    def dist_to(self, point):
        return distance(self.point, point)

    @property
    def flat(self):
#         import pdb; pdb.set_trace()
        yield self
        for c in self.children:
            for n in c.flat:
                yield n
    
    @property
    def ancestry(self):
        yield self
        if self.parent is not None:
            for a in self.parent.ancestry:
                yield a
                
    def attach(self, tree):
        self.children.append(tree)
        tree.parent = self
        
    def extend(self, goal, max_dist):
        d = distance(self.point, goal)
        if d <= max_dist:
            new_point = goal
        else:
            new_point = np.add(np.subtract(goal, self.point) * max_dist / d, self.point)
        return new_point
        
    def draw(self, ax, *args, **kwargs):
        for c in self.children:
            ax.plot([self.point[0], c.point[0]], [self.point[1], c.point[1]],*args, **kwargs)
            c.draw(ax, *args, **kwargs)
```
