```python
import random
import numpy as np
```

```python
xlim = [-10, 10]
ylim = [-10, 10]
def random_state_xy():
    return (random.uniform(*xlim), random.uniform(*ylim))

def closest_node(point, nodes):
    min_dist = np.inf
    best_node = None
    for node in nodes:
        d = node.dist_to(point)
        if d < min_dist:
            best_node = node
            min_dist = d
    return best_node

def state_check(point):
    return True
```

```python
class Tree(object):
    def __init__(self, point, children=None):
        self.point = np.array(point)
        
        if children is None:
            children = [] # unsafe to set this default in args, since appending to it later will affect the default for all future calls
            
        self.children = children
        
    def dist_to(self, point):
        return np.sqrt(np.sum(np.power(self.point - point, 2)))

    @property
    def flat(self):
#         import pdb; pdb.set_trace()
        yield self
        for c in self.children:
            for n in c.flat:
                yield n
                
    def attach(self, tree):
        self.children.append(tree)
        
    def draw(self, ax, **kwargs):
        for c in self.children:
            ax.plot([self.point[0], c.point[0]], [self.point[1], c.point[1]], **kwargs)
            c.draw(ax, **kwargs)
```
