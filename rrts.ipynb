```python
import random
import numpy as np
```

```python
xlim = [-10, 10]
ylim = [-10, 10]

def distance(p0, p1):
    return np.sqrt(np.sum(np.power(np.subtract(p0, p1), 2)))

def random_state_xy():
    return (random.uniform(*xlim), random.uniform(*ylim))

def closest_node(point, nodes):
    min_dist = np.inf
    best_node = None
    for node in nodes:
        d = node.dist_to(point)
        if d < min_dist:
            best_node = node
            min_dist = d
    return best_node


def state_check(point):
    return True
```

```python
class Tree(object):
    def __init__(self, point, children=None, parent=None):
        self.point = np.array(point)
        
        if children is None:
            children = [] # unsafe to set this default in args, since appending to it later will affect the default for all future calls
            
        self.children = children
        self.parent = parent
        
    def dist_to(self, point):
        return distance(self.point, point)

    @property
    def flat(self):
#         import pdb; pdb.set_trace()
        yield self
        for c in self.children:
            for n in c.flat:
                yield n
    
    @property
    def ancestry(self):
        yield self
        if self.parent is not None:
            for a in self.parent.ancestry:
                yield a
                
    def attach(self, tree):
        self.children.append(tree)
        tree.parent = self
        
    def extend(self, goal, max_dist):
        d = distance(self.point, goal)
        if d <= max_dist:
            return goal
        else:
            return np.add(np.subtract(goal, self.point) * max_dist / d, self.point)
        
    def draw(self, ax, *args, **kwargs):
        for c in self.children:
            ax.plot([self.point[0], c.point[0]], [self.point[1], c.point[1]],*args, **kwargs)
            c.draw(ax, *args, **kwargs)
```
